```python
import numpy as np
import pylab
import random
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
n = 300 
x = np.zeros(n)
y = np.zeros(n)
z = np.zeros(n)
for i in range(1, n):
    val = random.randint(1, 6)
    if val == 1:
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1]
        z[i] = z[i - 1]
    elif val == 2:
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1]
        z[i] = z[i - 1]
    elif val == 3:
        x[i] = x[i - 1]
        y[i] = y[i - 1] + 1
        z[i] = z[i - 1]
    elif val == 4:
        x[i] = x[i - 1]
        y[i] = y[i - 1] - 1
        z[i] = z[i - 1]
    elif val == 5:
        x[i] = x[i - 1]
        y[i] = y[i - 1] 
        z[i] = z[i - 1] + 1
    else:
        x[i] = x[i - 1]
        y[i] = y[i - 1]
        z[i] = z[i - 1] - 1
```

```python
def r_square(k):
    i=0
    rk=0
    while i<(n-k):
          rk += ((x[i+k]-x[i])**2+(y[i+k]-y[i])**2)/(n-k)
          i=i+1
    return rk

r = []
j=1
while j<n:
    r.append(r_square(j))
    j=j+1

t = list(range(1,n))
plt.plot(t,r,".b")

def func(x, a):
    return a*x
def fit(func, x, y):
    popt, pcov = curve_fit(func, x,y)
    return popt[0], pcov
x2 = np.linspace(t[0], t[298], num=50)
fit00 = fit(func, t, r)
plt.plot(x2, func(x2, fit00[0]), "-b",label="2D")
plt.xlabel("Time")
plt.ylabel("<r^2>")
plt.suptitle("Random Walker 2D")
plt.title("<r^2> versus Time")
print(fit00)
plt.savefig("r^2 versus Time 2D")
```
