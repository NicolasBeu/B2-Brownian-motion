```python
A = pd.read_excel("Water1(1).xlsx")
T = A[A.columns[1]].values.tolist()
D = A[A.columns[3]].values.tolist()
V = A[A.columns[5]].values.tolist()
print(T,D,V)
```

```python
from scipy.optimize import curve_fit
```

```python
import numpy as np
```

```python
plt.xlabel("T [K]")
plt.ylabel("D["u"\u03bcm^2/s]")
plt.title("Diffusion coefficient versus Temperature")
plt.suptitle("Water 0,5"u"\u03bcm")
plt.grid(which='both')
plt.plot(T,D,".b")

def func(x, a):
    return a*x
def fit(func, x, y):
    popt, pcov = curve_fit(func, x,y)
    return popt[0], pcov
x = np.linspace(T[0], T[8], num=50)
fit00 = fit(func, T, D)
plt.plot(x, func(x, fit00[0]), "-b")
print(fit00)
#plt.savefig("Diffusion coefficient 1 versus Temperature")
```
